```python
X = data.iloc[:, 2:].values
y = data.iloc[:, 1].values

```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
y = labelencoder_X_1.fit_transform(y)

# Splitting the dataset into Traioning set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
# X_test = sc.fit(X_test)

```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout

```

```python
classifier = Sequential()

```

Units  = Number of neurins in the layers

input_dim = number of columns of teh dataset.

Output_dim = Number of outputs to be fed to the next layer.

init = The way in which the wights should be Provided to an ANN.


```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(units=16, activation='relu', input_dim=30))


# Adding the droput to preevent overfitting
classifier.add(Dropout(rate=0.1))

```

```python
# Adding the second layer Hidden layer
classifier.add(Dense(units=16, activation='relu'))

classifier.add(Dropout(rate=0.1))

```

```python
# Adding the output layer
classifier.add(Dense(units=1, activation='sigmoid'))

```

```python
# Compilling the ANN

classifier.compile(optimizer= 'adam', loss='binary_crossentropy', metrics=['accuracy'])

```

```python
# Fitting the ANN to the training set.
history = classifier.fit(X_train, y_train, batch_size=100, epochs=150)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix

# Assuming y_test contains the true labels and y_pred contains the predicted labels
cm = confusion_matrix(y_test, y_pred)
```

```python
y_train_pred = (classifier.predict(X_train) > 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score

test_accuracy = accuracy_score(y_test, y_pred)

train_accuracy = accuracy_score(y_train, y_train_pred)

print(f" Test accuracy: {test_accuracy: .2f}")
print(f" Train accuracy: {train_accuracy: .2f}")

```

The Model is over fitting, hence we ae implementing Regularzaition techniques

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=1.0)

lasso.fit(X_train, y_train)

y_pred_lasso = lasso.predict(X_test)
```

```python
lasso_accuracy = accuracy_score(y_test, y_pred_lasso)
print(lasso_accuracy)

```

Implementing Ridge Regularization

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

ridge_model = Ridge(alpha=1.0)

ridge_model.fit(X_train, y_train)

y_pred_ridge = ridge_model.predict(X_test)


```

```python
mse = mean_squared_error(y_test, y_pred_ridge)

print(mse)
```
